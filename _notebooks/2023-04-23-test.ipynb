```java
import java.util.*;

public class Book {
    // argument constructor title?
    public String title;

    // unique id
    public String id;
    private static long idCounter = 0;

    // book count
    public static int bookCount;

    // shelf life
    public static int shelfLife;

    // age that determines shelf life
    public int age;

    // # of checkouts
    public int checkouts;

    public Book (){
        this.title = null;
    }

    public Book (String title){
        this.title = title;
        this.id = createID();
    }
    
    public static synchronized String createID() {
        return String.valueOf(idCounter++);
    }

    public String toString() {
        return("Book Title: "+title+" \nBook ID: "+id);
    }

    // checking whether its still fit to be on shelves
    public static boolean checkShelfLife() {
        if (shelfLife < 50) {
            return true; // book's shelf life is expired
        } else {
            return false; // book is still fine
        }
    }

}

class Novel extends Book {
    
    public String author;
    
    public Novel (String title){
        this.title = title;
        this.id = createID();
    }

    public Novel (String title, String author, int checkouts){
        this.title = title;
        this.id = createID();
        this.author = author;
        this.checkouts = checkouts;
    }

    @Override
    public String toString() {
        return (super.toString()+
                "\nAuthor: "+author);
    }

    public int checkLife() {
        shelfLife = 100 - (checkouts * 2);
        return (shelfLife);
    }

}

class Textbook extends Book {
    
    public String publisher;

    public Textbook (String title){
        this.title = title;
        this.id = createID();
    }

    public Textbook (String title, String publisher, int age){
        this.title = title;
        this.id = createID();
        this.publisher = publisher;
        this.age = age;
    }

    @Override
    public String toString() {     
        return (super.toString()+
                "\nPublisher: "+publisher);
    }

    public int checkLife() {
        shelfLife = 100 - (age * 3);
        return (shelfLife);
    }

}

// driver class
public class Tester 
{
    public static void main(String args[]) 
    {
            Book first = new Book("Coding Stuff");
            Novel second = new Novel("Fiction", "Higgins", 27);
            Textbook third = new Textbook("Nonfiction", "Willoughby", 3);

            Book[] books = {    // Use Array initialization to add book
                new Book("Barron's Computer Science"),  // Set a new Book object as array element.
                new Book("Angels and Demons"),
                new Book("Lion, Witch, and a Wardrobe")
            };

            Book.bookCount = books.length;

            System.out.println(first.toString()+"\n");

            System.out.println(second.toString());
            System.out.println("Shelf Life: "+second.checkLife());
            System.out.println("It should come off the shelf: "+second.checkShelfLife()+"\n");

            System.out.println(third.toString());
            System.out.println("Shelf Life: "+third.checkLife());
            System.out.println("It should come off the shelf: "+third.checkShelfLife());

            System.out.println("Library Book Count: " + Book.bookCount);
        
    }
}
Tester.main(null);
```

```java
import java.util.*;

public class Book {

    // argument constructor title
    public String title;

    // unique id
    public int id;

    // book count
    public static int bookCount;

    // shelf life
    public static int shelfLife;

    // age that determines shelf life
    public int age;

    // # of checkouts
    public int checkouts;

    public Book (){
        this.title = null;
        this.id = 0;
    }

    public Book (String title, int id){
        this.title = title;
        this.id = id;
        bookCount++;
    }

    public String toString() {
        return("Book Title: "+title+" \nBook ID: "+id);
    }

    // checking whether its still fit to be on shelves
    public static boolean checkShelfLife() {
        if (shelfLife < 50) {
            return true; // book's shelf life is expired
        } else {
            return false; // book is still fine
        }
    }

}

class Novel extends Book {
    
    public String author;

    public Novel (String title, int id){
        this.title = title;
        this.id = id;
    }

    public Novel (String title, int id, String author, int checkouts){
        this.title = title;
        this.id = id;
        this.author = author;
        this.checkouts = checkouts;
    }

    @Override
    public String toString() {
        return (super.toString()+
                "\nAuthor: "+author);
    }

    public int checkLife() {
        shelfLife = 100 - (checkouts * 2);
        return (shelfLife);
    }

}

class Textbook extends Book {

    public String publisher;

    public Textbook (String title, int id){
        this.title = title;
        this.id = id;
    }

    public Textbook (String title, int id, String publisher, int age){
        this.title = title;
        this.id = id;
        this.publisher = publisher;
        this.age = age;
    }

    @Override
    public String toString() {     
        return (super.toString()+
                "\nPublisher: "+publisher);
    }

    public int checkLife() {
        shelfLife = 100 - (age * 3);
        return (shelfLife);
    }

}

// driver class
public class Tester 
{
    public static void main(String args[]) 
    {
            Book first = new Book("Coding Stuff", 10);
            Novel second = new Novel("Fiction", 11, "Higgins", 27);
            Textbook third = new Textbook("Nonfiction", 12, "Willoughby",3);
    
            System.out.println(first.toString()+"\n");

            System.out.println(second.toString());
            System.out.println("Shelf Life: "+second.checkLife());
            System.out.println("It should come off the shelf: "+second.checkShelfLife()+"\n");

            System.out.println(third.toString());
            System.out.println("Shelf Life: "+third.checkLife());
            System.out.println("It should come off the shelf: "+third.checkShelfLife());
        
    }
}
Tester.main(null);
```
